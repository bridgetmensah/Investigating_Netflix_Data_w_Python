## The data

| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like

#sleceting the relase year column
netflix_df["release_year"]

# finding out which of those rows are >= 1990 ans <=1999
#I need logical_and() from numpy
import numpy as np

condition =  np.logical_and(netflix_df["release_year"]>=1990 , netflix_df["release_year"]<=1999)

#using the condition varaible to subset the netflix dataframe
netflix_df[condition]

#creating a variable filtered to save our filted dataframe 

filtered = netflix_df[condition]

#finding the most frequent duration 
most_frequent_duration = filtered['duration'].mode()[0]  # Use mode() to get the most frequent value

# Save the result as an integer
duration = int(most_frequent_duration)

print(f"The most frequent movie duration in the 1990s is: {duration}")

#ploting the most frequent movie duration
plt.hist(filtered["duration"],20,edgecolor='black')
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()

# Finding the short action movie
short_movie_count = 0
for i in range(len(filtered)):
    if filtered["genre"].iloc[i] == "Action" and filtered["type"].iloc[i] == "Movie" and  filtered["duration"].iloc[i] < 90:
        short_movie_count += 1
print(f"The number of short action movies is {short_movie_count}")

```
